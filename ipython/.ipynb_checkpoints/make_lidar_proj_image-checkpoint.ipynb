```python
import os
import sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
sys.path.append(os.path.abspath('../'))
```

```python
from utils.lidar_process import open_lidar
```

```python
points_set, camera_coord = open_lidar(
     '../test_images/test_1_lidar.pkl',
     4, 4, lidar_mean=[0, 0, 0], lidar_std=[1, 1, 1])
```

```python
points_set[]
```

```python
plt.plot(points_set[:,1] * 32)
```

```python
rgb = Image.open('../test_images/test_1_img.png').convert('RGB')
```

```python
plt.imshow(rgb)
```

```python
rgb_np = np.array(rgb)
#rgb_np.shape
rgb_zero = np.zeros([320,480,1], dtype=np.uint8)
```

```python
x = np.ones((320,480))
y = np.ones((320,480))
z = np.ones((320,480))
```

```python
rows = np.floor(camera_coord[:, 1])
cols = np.floor(camera_coord[:, 0])
```

```python
cmap = plt.cm.get_cmap('gray', 256)
cmap = np.array([cmap(i) for i in range(256)])[:, :3] * 256

x[(rows.astype(int), cols.astype(int))] = points_set[:, 0]
y[(rows.astype(int), cols.astype(int))] = points_set[:, 1]
z[(rows.astype(int), cols.astype(int))] = points_set[:, 2]
```

```python
#plt.imshow(np.array(x),cmap='gray')
cv2.imwrite('x.png', x)
```

```python
cmap = plt.cm.get_cmap('gray', 256)
cmap = np.array([cmap(i) for i in range(256)])[:, :3] * 256

for i in range(camera_coord.shape[0]):
    depth = points_set[i, 2]
    color = cmap[int(depth), :]
# For Waymo example
    rgb_zero[camera_coord[i, 1], camera_coord[i, 0]] = color
```

```python
plt.imshow(rgb_zero)
cv2.imwrite('x.png', rgb_zero)
```

```python
cmap = plt.cm.get_cmap('gray', 1633)
cmap = np.array([cmap(i) for i in range(1633)])[:, :3] * 1633
cmap
```

```python
for i in range(camera_coord.shape[0]):
    X = points_set[i, 2]
    color = cmap[int(512.0 / X), :]
# For Waymo example
    rgb_np[camera_coord[i, 1], camera_coord[i, 0]] = color
```
